# Design procedure for mass flow rate of supply air

This procedure applies to typical office and classroom spaces ([Price Ind. 2016](https://www.priceindustries.com/content/uploads/assets/literature/engineering-guides/displacement-ventilation-engineering-guide.pdf), page J-12).

## Determine the summer or winter load of the thermal zone

The sensible load is ([T02_load_winter.ipynb](T02_load_winter.ipynb)):

$$\dot Q_s = (UA + \dot m_a c)(\theta_O - \theta_I)  + \dot{Q}_{s,a}$$

and the latent load is ([T02_load_winter.ipynb](T02_load_winter.ipynb)): 

$$\dot Q_l = \dot m_a l (w_O - w_I) + \dot{Q}_{l,a}$$

where:
- $UA = \sum_{i}f_i U_i S_i+ \sum_{i} \psi_i L_i$ - global conductivity of the building (walls, windows, thermal bridges, etc.), W/K;
    - $U_i$ is the [overall heat transfer coefficient](https://en.wikipedia.org/wiki/R-value_(insulation) of the wall $i$, W/(m²·K);
    - $S_i$ - surface area of the wall $i$, m²; 
    - $f_i$ - correction for heat loss of wall $i$;
    - $\psi_i$ - linear thermal transmittance of thermal bridge $i$, W/(m·K);
    - $L_i$ - length of the thermal bridge $i$, m;
- $\dot m_a = \dot{m}_{inf} + \dot m_v$ - mass flow rate of outdoor air, kg/s;
    - $\dot{m}_{inf}$ - mass flow rate by [infiltration](https://en.m.wikipedia.org/wiki/Infiltration_(HVAC)), kg/s;
    - $\dot{m_v}$ - mass flow rate by natural [ventilation](https://en.wikipedia.org/wiki/Ventilation_(architecture)), kg/s;
- $c$ - [specific heat capacity of the air](https://en.wikipedia.org/wiki/Specific_heat_capacity), J/(kg·K);
- $l$ - [specific latent heat](https://en.wikipedia.org/wiki/Latent_heat#Specific_latent_heat), J/kg;
- $\theta_O$ - outdoor temperature, °C;
- $\theta_I$ - indoor temperature, °C;
- $w_O$ - water content (humidity ratio) of the outdoor air, kg/kg;
- $w_I$ - water content (humidity ratio) of the indoor air, kg/kg;
- $\dot{Q}_{s,a}$ - sensible auxiliary heat gains (from occupants, electrical devices, solar radiation, etc.), W;
- $\dot{Q}_{l,a}$ - latent auxiliary heat gains (from occupants, evaporative processes like cooking, etc.), W.

*Note*: $HLC = UA + \dot m_a c$ is the overall *Heat Loss Coeffcient*, W/K ([Ghiaus, Alzetto, 2019](https://doi.org/10.1080/19401493.2018.1561753)).

## Chose the type of terminal units

A terminal unit is the element of the ventilation system that provides supply air to a room. This can be, for example, grilles, chilled beams, or other types of diffusers ([France air](https://www.france-air.com/guides-et-documents/regles-de-lart-en-diffusion/la-diffusion-par-type/), [Price Ind. 2014](https://www.priceindustries.com/content/uploads/assets/literature/engineering-guides/terminal-units-engineering-guide.pdf)).

Simple terminal units allow air to be supplied to the room without any particular features. They ensure diffusion by mixing the supplied air with ambient air.

Terminal units with reflection are designed to allow the air jet to adhere to the ceiling thanks to the Coanda effect. This allows for better diffusion and mixing of the supplied air with ambient air.

Critical terminal units are designed to operate at low speed and allow diffusion by air displacement. This allows for more effective pollutant removal and better air quality in the room.

The type of terminal is chosen considering many criteria, one of which is maximum admited thermal load.

## Determine the mass flow rate of supply air 
The mass flow rate of supply air is determined as a function of thermal comfort, minimum ventilation rate for indoor air quality, and minimum circulation rate for air mixing in the space.

![supply_path](./Figures/supply_air.svg)

>Figure 1 Model of the thermal zone or the supply air path


The [equations for the thermal zone](README.md) are:

\begin{align*}
    &\left\{
    \begin{aligned}
        &\dot m c \theta_I = \dot m c \theta_S + \dot Q_s\\
        &\dot m l w_I = \dot m l w_S + \dot Q_l
    \end{aligned}
    \right. \\
\end{align*}

where:
- $\dot m$ - mass flow rate of supply air, kg/s;
- $\theta_I, \theta_S$ - temperature of indoor air and of supply air, respectively, °C;
- $w_I, w_S$ - humidity ratio of indoor air and of supply air, respectively, kg/kg;
- $\dot Q_s, \dot Q_l$ - sensible and latent heat load, respectively, of the thermal zone, W;
- $c$ - [specific heat capacity](https://en.m.wikipedia.org/wiki/Specific_heat_capacity) of dry air, $c$ = 1⋅10³ J/(kg⋅K);
- $l$ - [specific latent heat](https://en.m.wikipedia.org/wiki/Latent_heat#Specific_latent_heat) of water vapor, $l$ = 2496⋅10³ J/kg.

Let's consider that the indoor conditions $\theta_I$ and $w_I$ are imposed and the loads $\dot Q_s$ and $\dot Q_l$ are known. Then, this system of two energy balance equations has three unknowns: $\dot m$, $\theta_S$, and $w_S$. On the psychrometric chart, it is represented by the *supply air line* that connects points *I* and *S*. Any point *S* on this line satisfies the energy balance equations (for sensible and latent heat). The position of point *S* depends on the mass flow rate of supply air, $\dot m$. This mass flow rate is determined from (Del Zotto et al. 2003, pp.152, 188, Price Ind. 2016, p.J-12):
- _thermal comfort_: temperature differece between supply air and indoor air;
- _indoor air quality_: minimum outdoor (fresh) air flow rate as a function of number of occupants;
- _air mixing_: minimum supply air flow rate as a function of the characteristics of air terminals.

It is solved by imposing $\theta_S$ of by imposing $\dot m$.

### Air mixing

A sufficient air change rate is necessary to ensure the mixing and circulation of air within the ventilated space. For air-conditioned thermal zones, a [circulation rate](https://www.xpair.com/lexique/definition/taux_de_brassage.htm) of 8 vol/h is an usual value; the value of 4 vol/h is considered minimal. Recommended values for circulation rates are given in Table 5.
